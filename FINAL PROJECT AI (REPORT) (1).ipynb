# INTRODUCTION

The Vision Transformer (ViT) is an innovative deep learning model designed to process visual data using the Transformer architecture, initially developed for natural language processing (NLP). Unlike traditional Convolutional Neural Networks (CNNs), which rely on convolutional layers to extract features from images, ViT introduces a paradigm shift by leveraging the Self-Attention mechanism. This allows the model to understand the relationships between different regions of an image, regardless of their spatial distance. This unique approach opens up new possibilities for computer vision tasks, such as image classification, object detection, and semantic segmentation[1], where understanding both local and global relationships is crucial.
One of the key advantages of ViT is its ability to generalize well across various visual tasks. However, despite its impressive accuracy, the model has some limitations. The self-attention mechanism, while powerful, is computationally expensive and requires significant memory resources. This makes traditional ViT architectures less practical for real-time or mobile applications, where efficiency is just as important as accuracy.
To address these limitations, a new hybrid architecture called FastViT [2] has recently emerged. FastViT combines the strengths of transformers and convolutional networks, representing a breakthrough in efficiency and speed. This architecture incorporates RepMixer, an innovative component that restructures the architecture by removing multi-path connections, significantly reducing memory access costs. By using structural reparameterization, FastViT achieves faster processing without compromising accuracy.
In addition to RepMixer, FastViT employs other optimization techniques such as train-time overparameterization and large kernel convolutions. These improvements make it highly competitive for a variety of computer vision tasks, including image classification, object detection, segmentation, and even 3D mesh regression. Compared to traditional models like EfficientNet and ConvNeXt, FastViT demonstrates faster inference times while achieving superior or comparable accuracy.
The applications of both ViT and FastViT are numerous and inspiring in the field of computer vision. One of the most prominent use cases is in autonomous driving, where real-time segmentation and object detection are essential for safe navigation. FastViT's efficiency makes it well-suited for such environments where quick decision-making is critical. Another key application is in facial recognition, with Apple's Face ID being a well-known example. By analyzing facial features precisely, models like ViT contribute to the enhanced security of smartphones and digital systems.
Moreover, the role of vision transformers extends to Augmented and Virtual Reality (AR/VR) technologies. These models enhance virtual worlds by accurately recognizing and interacting with the environment, creating a more immersive and interactive user experience. FastViT's speed and robustness make it an excellent candidate for enhancing AR/VR applications where low latency is essential.
In conclusion, ViT and its successor, FastViT, showcase how innovation in deep learning can reshape computer vision. By tackling the challenges of computational inefficiency, FastViT paves the way for practical, real-time applications while maintaining state-of-the-art accuracy across various visual tasks. Their potential applications in autonomous systems, security, and immersive technologies continue to highlight their transformative impact on the world of computer vision.



# Methode

The FastViT model is a hybrid vision transformer that integrates the strengths of Transformers and Convolutional Neural Networks (CNNs) to achieve a state-of-the-art trade-off between latency and accuracy. The architecture is designed based on three key principles:
1-RepMixer Block:
A novel token mixer that removes skip connections using structural reparameterization to reduce memory access costs and improve computational efficiency.
2-Linear Train-Time Overparameterization:
Additional branches are introduced during training to increase model capacity. These branches are reparametrized during inference to reduce computational overhead.
3-Large Kernel Convolutions:
Instead of relying solely on self-attention for information mixing, large kernel convolutions are employed in the early stages to enhance receptive fields while maintaining low latency.
The model is divided into multiple stages where RepMixer is primarily used for token mixing, and self-attention layers are selectively incorporated for later stages requiring global feature extraction.
FastViT architecture focuses on improving the latency-accuracy tradeoff in vision transformer models for image processing tasks like classification, detection, and segmentation. Conventional Vision Transformers (ViTs) offer strong performance but are computationally intensive, with significant memory and latency overheads. By integrating convolutional principles and introducing architectural optimizations, FastViT aims to make ViT models both faster and more efficient without sacrificing accuracy.
Problem:
Traditional Vision Transformers (ViTs), despite their impressive accuracy, suffer from high computational and memory costs due to the self-attention mechanism, especially when processing high-resolution images. These issues lead to longer inference times and higher power consumption, making ViTs inefficient for mobile and embedded devices.
Solution:
The FastViT team tackled these issues through the following innovations:
1-Eliminating Skip Connections:
By introducing RepMixer, skip connections (which often increase memory access latency) were removed. This architectural change significantly reduced latency while maintaining high accuracy.
2-Efficient Convolutional Operations:
The use of factorized convolutions (depth-wise followed by pointwise convolutions) further optimized computational efficiency.
3-Train-Time Overparameterization:
By introducing overparameterization during training, the capacity of factorized layers was enhanced without sacrificing efficiency during inference.
4-Large Kernel Convolutions:
Incorporating large kernel convolutions instead of self-attention in early stages maintained accuracy while reducing latency.

FastViT Tasks:
1. Image Classification 
-Task: Classifying images into one of many categories.
Dataset: ImageNet-1K dataset-
Size: ~1.3 million training images and 50,000 validation images.-
-Performance: FastViT achieved up to 83.9% Top-1 accuracy, outperforming EfficientNet and ConvNeXt with significantly lower latency.
2. Object Detection 
-Task: Identifying and locating objects within images by drawing bounding boxes around them.
-Dataset: MS-COCO dataset
-Size: 118,000 training images and 5,000 validation images with 80 object classes.
-Performance: FastViT was 4.3x faster on mobile devices and 2.4x faster on desktop GPUs compared to models like CMT-S, achieving similar or superior detection accuracy.
3. Semantic Segmentation 
Task: Classifying each pixel in an image to its corresponding object category.-
Dataset: ADE20K dataset-
-Size: 20,000 training images and 2,000 validation images with 150 semantic categories.
-Performance: FastViT-MA36 achieved a 5.2% higher mIoU (Mean Intersection over Union) than PoolFormer-M36, with lower FLOPs, parameter count, and latency.
4. 3D Hand Mesh Estimation 
Task: Estimating the 3D structure of a hand from a single 2D image.-
-Dataset: FreiHAND dataset
Size: 32,560 training samples and 3960 testing samples.-
-Performance: FastViT-MA36 outperformed state-of-the-art real-time models such as MobileHand and MobRecon, being 1.9x faster than MobileHand and 2.8x faster than MobRecon, while delivering superior accuracy for joint and vertex estimation.

# Conclusion

The FastViT model represents a significant advancement in computer vision by combining the strengths of transformers and convolutional neural networks (CNNs) to achieve state-of-the-art performance with minimal latency. This hybrid architecture addresses long-standing challenges in the field, such as the computational and memory overhead associated with traditional vision transformers, making it highly efficient for deployment on both mobile devices and desktop GPUs.
One of the most innovative components of FastViT is the RepMixer block, which leverages structural reparameterization to eliminate skip connections, significantly reducing memory access costs without sacrificing performance. Additionally, the adoption of linear train-time overparameterization further boosts the capacity of the network during training while keeping inference computationally efficient. By incorporating large kernel convolutions, FastViT manages to efficiently capture information in the early stages of the network, providing a competitive alternative to self-attention mechanisms that are typically computationally expensive.
Through comprehensive experiments, the research demonstrated that FastViT excels in multiple tasks such as image classification, object detection, semantic segmentation, and 3D hand mesh estimation. On the ImageNet-1K dataset, FastViT achieved an impressive Top-1 accuracy of 83.9%, outperforming major competitors like EfficientNet and ConvNeXt with significantly faster inference speeds. In object detection, the model proved to be 4.3x faster on mobile devices compared to CMT-S while maintaining similar accuracy. For semantic segmentation on the ADE20K dataset, FastViT-MA36 achieved a 5.2% higher mIoU than PoolFormer-M36. Furthermore, in 3D hand mesh estimation, it outperformed models like MobileHand and MobRecon by a significant margin, being both faster and more accurate.
Another noteworthy achievement of FastViT is its robustness to out-of-distribution samples and corruptions, making it suitable for real-world applications where variability in input data is common. Its efficient design, scalability, and superior performance across tasks make it a versatile and powerful solution for modern computer vision applications. This work paves the way for future innovations in hybrid models, blending the best of transformers and convolutional techniques to push the boundaries of what is possible in computer vision.
